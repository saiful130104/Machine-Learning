<h2>Matplotlib Tutorial</h2>

```python
import matplotlib.pyplot as plt

%matplotlib inline
import numpy as np
```

```python
x = np.arange(0,10)
y = np.arange(11,21)
```

<h3>Plotting using matplotlib</h3>

<h4>Using scatter</h4>

```python
# plt scatter
plt.scatter(x,y,c='r',s=50)

# Graph title
plt.title('Graph in 32D')

# Title of x-axis
plt.xlabel('X axis')

# Title of Y-axis
plt.ylabel('Y axis')

# To save the graph (Filename)
plt.savefig('Test.png')
```

<h4>Using plot</h4>

```python
# plt plot

y = x*x

plt.plot(x,y,'g*--',linewidth=2, markersize=12)

# Graph title
plt.title('Graph in 32D')

# Title of x-axis
plt.xlabel('X axis')

# Title of Y-axis
plt.ylabel('Y axis')

# To save the graph (Filename)
plt.savefig('Test.png')
```

<h3>Subplot</h3>

```python
# Multiple plotting using Subplot 

plt.subplot(2,2,1)
plt.title('First')
plt.plot(x,y,'r--',marker='*')

plt.subplot(2,2,2)
plt.title('Second')
plt.plot(x,y,'g--',marker='o')

plt.subplot(2,2,3)
plt.title('Third')
plt.plot(x,y,'b--',marker='*')

plt.subplot(2,2,4)
plt.title('Fourth')
plt.plot(x,y,'y--',marker='o')
```
